```python
from PIL import Image
import requests
from io import BytesIO
import IPython.display as display

def display_image(url) :
  response = requests.get(url)
  img = Image.open(BytesIO(response.content))

  display.display(img)
```

```python
response = openai.Image.create(
    prompt="An image of a rabbit in the forest, high quality, the Simpsons TV style",
    size='512x512', # '256x256', '512x512', '1024x1024'
    n=2
)

display_image(response['data'][0]['url'])
display_image(response['data'][1]['url'])
```

```python
# 이미지 변형

response = openai.Image.create_variation(
    image=open("image.png", "rb"),
    size="512x512",
    n=2
)

display_image(response['data'][0]['url'])
display_image(response['data'][1]['url'])
```

```python
# 이미지 특정 부분 수정

response = openai.Image.create_edit(
    image=open("original.png", "rb"),
    mask=open("mask-2.png", "rb"),
    prompt="An image of a happy rabbit in the forest, high quality, the Simpsons TV style, background is in the middle of night",
    size="512x512",
    n=2
)

display_image(response['data'][0]['url'])
display_image(response['data'][1]['url'])
```
